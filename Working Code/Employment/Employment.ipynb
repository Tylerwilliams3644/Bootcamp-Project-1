```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
naics_path = "../../Resources/employment_by_NAICS.csv"
naics_df = pd.read_csv(naics_path, encoding="ISO-8859-1")

display(naics_df)
```

```python
naics_df = naics_df[["GeoName", "Description", "Unit", "2012", "2013", "2014", "2015", "2016", "2017", "2018", "2019", "2020", "2021", "2022"]]
cleaned_naics_df = naics_df.rename(columns={"GeoName" : "Location"})
cleaned_naics_df
```

```python
employment_df = cleaned_naics_df.loc[cleaned_naics_df["Unit"] == "Number of jobs"]
display(employment_df)
```

```python
desired_states = ["Alabama", "Alaska", "Arizona", "Arkansas","California","Colorado","Connecticut","Delaware","Florida","Georgia","Hawaii","Idaho","Illinois","Indiana","Iowa","Kansas","Kentucky","Louisiana","Maine","Maryland","Massachusetts","Michigan","Minnesota","Mississippi","Missouri","Montana","Nebraska","Nevada","New Hampshire","New Jersey","New Mexico","New York","North Carolina","North Dakota","Ohio","Oklahoma","Oregon","Pennsylvania","Rhode Island","South Carolina","South Dakota","Tennessee","Texas","Utah","Vermont","Virginia","Washington","West Virginia","Wisconsin","Wyoming"]

employment_by_state = employment_df.loc[employment_df["Location"].isin(desired_states)]
employment_by_state = employment_by_state.reset_index(drop = True)
employment_by_state["Description"] = employment_by_state["Description"].str.strip()


employment_by_state.head(33)
```

```python
employment_by_state["Description"].unique()
```

Government Jobs

```python
government_jobs = employment_by_state.loc[employment_by_state["Description"].isin(["Local government", "State government", "State and local", "Military", "Federal civilian", "Government and government enterprises"])]

government_jobs.tail(20)
```

```python
desired_years = "2012", "2013", "2014", "2015", "2016", "2017", "2018", "2019", "2020", "2021", "2022"

for year in desired_years:
    government_jobs[year] = pd.to_numeric(government_jobs[year], errors='coerce')

desired_states = ["Alabama", "Alaska", "Arizona", "Arkansas","California","Colorado","Connecticut","Delaware","Florida","Georgia","Hawaii","Idaho","Illinois","Indiana","Iowa","Kansas","Kentucky","Louisiana","Maine","Maryland","Massachusetts","Michigan","Minnesota","Mississippi","Missouri","Montana","Nebraska","Nevada","New Hampshire","New Jersey","New Mexico","New York","North Carolina","North Dakota","Ohio","Oklahoma","Oregon","Pennsylvania","Rhode Island","South Carolina","South Dakota","Tennessee","Texas","Utah","Vermont","Virginia","Washington","West Virginia","Wisconsin","Wyoming"]

all_government_jobs = []

for state in desired_states:
        find_government_job = government_jobs.loc[government_jobs["Location"] == state, desired_years].sum()
        all_government_jobs.append(find_government_job)

all_government_jobs = pd.DataFrame(all_government_jobs)
all_government_jobs["State"] = desired_states

column_to_move = "State"
moved_column = all_government_jobs.pop(column_to_move)
all_government_jobs.insert(0,column_to_move, moved_column)
display(all_government_jobs)
```

```python
all_government_jobs['2022'] = pd.to_numeric(all_government_jobs['2022'])
all_government_jobs.plot.bar(x='State', y='2022', figsize = (10,5))
plt.ticklabel_format(style="plain", axis="y")
plt.xlabel("States")
plt.ylabel("Number of Jobs")
plt.title("Number of Government Jobs in 2022")
```
