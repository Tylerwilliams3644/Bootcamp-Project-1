```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
naics_path = "../../Resources/employment_by_NAICS.csv"
naics_df = pd.read_csv(naics_path, encoding="ISO-8859-1")

display(naics_df)
```

```python
naics_df = naics_df[["GeoName", "Description", "Unit", "2012", "2013", "2014", "2015", "2016", "2017", "2018", "2019", "2020", "2021", "2022"]]
cleaned_naics_df = naics_df.rename(columns={"GeoName" : "Location"})
cleaned_naics_df
```

```python
employment_df = cleaned_naics_df.loc[cleaned_naics_df["Unit"] == "Number of jobs"]
display(employment_df)
```

```python
desired_states = ["Alabama", "Alaska", "Arizona", "Arkansas","California","Colorado","Connecticut","Delaware","Florida","Georgia","Hawaii","Idaho","Illinois","Indiana","Iowa","Kansas","Kentucky","Louisiana","Maine","Maryland","Massachusetts","Michigan","Minnesota","Mississippi","Missouri","Montana","Nebraska","Nevada","New Hampshire","New Jersey","New Mexico","New York","North Carolina","North Dakota","Ohio","Oklahoma","Oregon","Pennsylvania","Rhode Island","South Carolina","South Dakota","Tennessee","Texas","Utah","Vermont","Virginia","Washington","West Virginia","Wisconsin","Wyoming"]

employment_by_state = employment_df.loc[employment_df["Location"].isin(desired_states)]
employment_by_state = employment_by_state.reset_index(drop = True)


employment_by_state.head(50)
```
